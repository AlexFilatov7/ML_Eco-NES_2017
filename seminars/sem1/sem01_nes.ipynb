### Предсказать кассовые сборы: регрессия

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
df_dropna = df.dropna()
X = df_dropna[['duration', 'budget', 'aspect_ratio', 
                   'director_facebook_likes', 'actor_1_facebook_likes']]
y = df_dropna.gross
```

```python
regression.fit(X, y)
```

```python
regression.score(X, y)
# так на самом деле не делают — проверка модели на тех же данных, на которых
# мы её обучали — не очень хорошая стратегия, т.к. модель может хорошо
# работать на тех данных, на которых её обучили, но плохо на других данных
# (это называется overfitting, переобучение)
```

```python
regression.coef_
```

### Предсказать, окупится ли фильм?

Окупился ли фильм? Считая, что кинотеатр забирает себе половину…

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
y = df_dropna.gross > df_dropna.budget * 2
classifier.fit(X, y)
```

```python
classifier.coef_
```

```python
classifier.score(X, y)
```

```python
df.assign(pay_for_itself = lambda x: x.gross > x.budget*2)
```

### Как добавить категориальные данные в модель

```python
content_ratings = set(df_dropna.content_rating)
```

```python
df_content_ratings = pd.concat([df_dropna, 
                                pd.get_dummies(df_dropna.content_rating)], 
                               axis=1)
```

```python
X = df_content_ratings[['duration', 'budget', 'aspect_ratio', 
                   'director_facebook_likes', 'actor_1_facebook_likes'] + 
                       list(content_ratings)]
```

```python
classifier.fit(X, y)
```

```python
classifier.score(X, y)
```
